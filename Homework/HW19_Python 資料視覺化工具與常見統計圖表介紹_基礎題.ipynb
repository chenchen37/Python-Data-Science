```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

plt.plot(x,y)
plt.show() 

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)
plt.show()

x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y, X)
plt.scatter(X, Y, alpha = 0.5, s = 50, c = T)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
